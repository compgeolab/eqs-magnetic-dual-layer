```python
fig = pygmt.Figure()
fig.coast(region=[-180, 180, -90, -60], projection="S0/-90/12c", frame="afg", land="gray")

# Need pygmt.makecpt to plot colour.
# cmap="polar+h" -> +h makes the sure the centre is at zero (white = 0).
pygmt.makecpt(cmap="polar+h", series=[icegrav.MagL.min(), icegrav.MagL.max()])

fig.plot(
    x=icegrav.MagL,
    y=icegrav.MagL,
    color=(icegrav.MagL),
    style="c0.02c",
    cmap=True,
)
fig.colorbar(position="JBC")
fig.show()
```
